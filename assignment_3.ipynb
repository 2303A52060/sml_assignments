```python
#step-2
from functools import lru_cache

class Game:
    def __init__(self, initial):
        self.initial = initial

    def actions(self, state):
        """Return a collection of the allowable moves from this state."""
        raise NotImplementedError

    def result(self, state, move):
        """Return the state that results from making a move from a state."""
        raise NotImplementedError

    def is_terminal(self, state):
        """Return True if this is a final state for the game."""
        return not self.actions(state)

    def utility(self, state, player):
        """Return the value of this final state to player."""
        raise NotImplementedError
#step-3

class TicTacToe(Game):
    def __init__(self):
        super().__init__(initial=self.initial_state())

    def initial_state(self):
        """Return the initial state of the Tic-Tac-Toe game."""
        return [[' ' for _ in range(3)] for _ in range(3)]

    def actions(self, state):
        """Return a list of possible moves (empty cell positions)."""
        return [(r, c) for r in range(3) for c in range(3) if state[r][c] == ' ']

    def result(self, state, move):
        """Return the state resulting from making a move."""
        new_state = [row[:] for row in state]
        r, c = move
        current_player = 'X' if sum(row.count('X') for row in state) <= sum(row.count('O') for row in state) else 'O'
        new_state[r][c] = current_player
        return new_state

    def is_terminal(self, state):
        """Check if the game is over."""
        return self.utility(state, 'X') is not None or self.utility(state, 'O') is not None or all(state[r][c] != ' ' for r in range(3) for c in range(3))

    def utility(self, state, player):
        """Return the utility of the state for the given player."""
        # Check rows, columns and diagonals
        for r in range(3):
            if all(state[r][c] == player for c in range(3)):
                return 1 if player == 'X' else -1
        for c in range(3):
            if all(state[r][c] == player for r in range(3)):
                return 1 if player == 'X' else -1
        if all(state[i][i] == player for i in range(3)):
            return 1 if player == 'X' else -1
        if all(state[i][2 - i] == player for i in range(3)):
            return 1 if player == 'X' else -1
        if all(state[r][c] != ' ' for r in range(3) for c in range(3)):
            return 0
        return None
#step-4
def play_game(game, strategies, verbose=False):
    """Play a turn-taking game. 'strategies' is a {player name: function} dict."""
    state = game.initial
    while not game.is_terminal(state):
        player = 'X' if sum(row.count('X') for row in state) <= sum(row.count('O') for row in state) else 'O'
        move = strategies[player](game, state)
        state = game.result(state, move)
        if verbose:
            print('Player', player, 'move:', move)
            for row in state:
                print(row)
    return 0
#example strategy function
import random

def random_strategy(game, state):
    return random.choice(game.actions(state))
#example usage
if __name__ == "__main__":
    tic_tac_toe = TicTacToe()
    strategies = {'X': random_strategy, 'O': random_strategy}
    play_game(tic_tac_toe, strategies, verbose=True)


```
